```python
import numpy as np
```

```python
# Calcular las raices del polinomio 1 - 0.3x + 0.4x^2 - 0.1x^3 - 0.9x^4
roots = np.roots([-0.9, -0.1, 0.4, 0.3, 1])
abs_roots = np.abs(roots)
print("Ejercicio 3")
print(roots, "\n")
print(abs_roots, "\n")
```
